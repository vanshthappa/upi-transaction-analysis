```python
# Since my project is UPI-focused
upi_df = df[df['description'].str.contains('upi')]
```

```python
# Function for categorization of payments
def categorize(desc):
    if any(x in desc for x in ['swiggy', 'zomato', 'dominos', 'eatclub', 'mcdonald']):
        return 'food'
    elif any(x in desc for x in ['uber', 'ola', 'rapido', 'metro']):
        return 'travel'
    elif any(x in desc for x in ['amazon', 'flipkart', 'myntra', 'zepto', 'blinkit']):
        return 'shopping'
    elif any(x in desc for x in ['recharge', 'jio', 'airtel']):
        return 'bills'
    elif any(x in desc for x in ['salary']):
        return 'income'
    else:
        return 'others'
```

```python
# Pass the description column in the function
upi_df['category'] = upi_df['description'].apply(categorize)
```
